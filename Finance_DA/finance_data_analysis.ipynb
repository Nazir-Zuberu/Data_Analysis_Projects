# Analyzing AAPL

## Daily price and price change

```python
df = pd.read_csv(path + '/' + "AAPL_data.csv")
df.head()
```

```python
df['daily_price_change'] = df.close - df.open
df.head()
```

```python
df['daily_percentage_change'] = ((df.close - df.open)/df.close)*100
df.head()
```

```python
px.line(df, x = 'date', y = 'daily_percentage_change')
```

```python
df['date'] = pd.to_datetime(df.date)
```

```python
df = df.set_index('date')
df.head()
```

```python
df.close.resample("M").mean().plot() # Resampling on monthly basis
```

```python
df.close.resample("Y").mean().plot(kind = 'bar') # Resampling on annual basis
```

```python
df.resample("Y")['close'].mean()
```

# Relationship among companies

```python
close = pd.DataFrame()
for company in company_list:
    df = pd.read_csv(path + '/' + company+"_data.csv")
    close[company] = df.close
```

```python
close.head()
```

```python
#!pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.pairplot(close)
```

```python
sns.heatmap(close.corr(), annot=True)
```

## Relationship among percentage change in price of companies

```python
rel_Price = pd.DataFrame()
for company in company_list:
    df = pd.read_csv(path + '/' + company+"_data.csv")
    rel_Price[company] = ((df.close - df.open)/df.close)*100
```

```python
rel_Price.head()
```

```python
sns.pairplot(rel_Price)
```

```python
sns.heatmap(rel_Price.corr(), annot=True)
```
